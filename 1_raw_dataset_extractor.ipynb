```python
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

import scipy.io
from mat73 import loadmat
import numpy as np
import csv

import os
import time

# Function to read a .mat file and write to .csv
def convertToCsv(index):
    raw_file_name = 'raw_dataset/S00' + str(index) + '_tr.mat'

    print("Processing file: ", raw_file_name)
    
    # Load the .mat file using mat73 library
    data = loadmat(raw_file_name)

    # Keep only the required variable.
    data = data['train_split_sub_wise_data']

    # Convert the variable to a numpy array
    array = np.array(data)
    headers = ['subject_id', 'index', 'semg_1', 'semg_2', 'semg_3', 'semg_4', 'semg_5', 'sub_activity', 'activity', 'repetition_index', 'trial_index', 'manual_sub_activity']

    # Specify the output CSV file path
    output_file_path = 'dataset_csv_stage/indexed_all_subjects.csv'

    if os.path.exists(output_file_path):
        os.remove(output_file_path)

    output_file_path = 'dataset_csv_stage/indexed_S00'
    with open(output_file_path + str(index) + '.csv', 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(headers)
        
        counter = 1
        current_activity = 1
        for i, row in enumerate(array, start=1):
            if current_activity != int(list(row)[6]):
                print("Activity changed to ", int(list(row)[6]), "...counter resetting..")
                current_activity = int(list(row)[6])
                counter = 1
            writer.writerow([index] + [counter] + list(row))
            counter += 1
    print("Completed processing file: ", raw_file_name)
```

```python
# Index numbers list of the raw dataset .mat files
file_index_list = [1, 2, 5, 6, 7, 8, 9, 10]

# Get the start time
st = time.time()

for index in file_index_list:
    convertToCsv(index)

# Get the end time
et = time.time()

# Calculate the elapsed time
elapsed_time = et - st
print('Elapsed time:', elapsed_time, 'seconds')
```
